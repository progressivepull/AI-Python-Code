# Check required columns

```python
required_columns = [
    "ID", "Name", "State", "Height", "Weight",
    "Hair Length", "Facial Age Estimate",
    "Education Level", "ID Type", "Income"
]
```

```python
# 1. Count the number of columns
num_cols = len(df.columns)
print(f"Total columns found: {num_cols}")

# 2. Check for required columns and track status
all_found = True
for col in required_columns:
    if col not in df.columns:
        print(f"Missing required column: {col}")
        all_found = False
    else:
        print(f"Column found: {col}")

# 3. Final confirmation
if all_found:
    print("Success: All required columns are present.")
else:
    print("Column is missing")
```
